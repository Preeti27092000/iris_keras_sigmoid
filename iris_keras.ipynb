```python
df_set = df[df['Species'] == 'Iris-setosa']
df_set
```

```python
df['Species'] == 'Iris-setosa'
```

```python
df_vers = df[df['Species'] == 'Iris-versicolor']
df_vers
```

```python
df_merge = pd.merge(df_set,df_vers,how = 'outer')
```

```python
df_merge
```

```python
df_cat = df_merge.select_dtypes(include = object)

df_cat['Species'].replace(to_replace = ['Iris-setosa'],value = 0,inplace = True)
df_cat['Species'].replace(to_replace = ['Iris-versicolor'],value = 1,inplace = True)
```

```python
df_num = df_merge.select_dtypes(include = 'number')
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
scale = ss.fit_transform(df_num)
df_scaled = pd.DataFrame(scale,columns = df_num.columns)
df_scaled.head()
df_scaled.drop('Id',inplace = True,axis = 1)
```

```python
df_concat = pd.concat([df_scaled,df_cat],axis = 1)
df_concat
```

```python
target = df_merge['Species']
df_new = df_concat.drop(['Species'],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_new,target,random_state = 1,test_size = 0.3)
```

```python
y_train = y_train.astype(int)
```

```python
y_train.dtypes
```

```python
y_test = y_test.astype(int)
```

```python
from tensorflow.keras import layers
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(3,input_shape = (4,),activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(optimizer = 'Adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs = 100) 
```

```python
model.evaluate(x_test,y_test)
```
